```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
y = np.linspace(10,40, 7)/100

h2 = 2872.06
h7 = 1087.4
h5 = 201.51
h3 = 2037.04
h1 = 3157.9

h6 = y*(h2-h7) + h5
qa = h1-h6
wt = h1-h2 + (1-y)*(h2-h3)

plt.plot(y, h6, 'r-', marker='s', label='entalpia do ponto 6')

# Configurações do gráfico
plt.xlabel('y (Fraçao de extração)', fontsize=12)
plt.ylabel('Entalpia (kJ/kg)', fontsize=12)
plt.title('h6 para diferentes extrações', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

print(y)
print(h6)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, qa, 'b-', marker='o', label='qa (Calor adicionado)')
plt.plot(y, wt, 'r-', marker='s', label='wt (Trabalho da turbina)')
plt.xlabel('y (Fraçao de extração)', fontsize=12)
plt.ylabel('Energia (kJ/kg)', fontsize=12)
plt.title('Variação de calor adicionado e trabalho da turbina em função de y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

plt.show()
print(qa)
print(wt)
```

```python
eficiencia = (wt - 9.7)/qa*100

plt.plot(y, eficiencia, 'y-', marker='s')
plt.xlabel('y (Fraçao de extração)', fontsize=12)
plt.ylabel('Eficiência', fontsize=12)
plt.title('Variação da eficiência em função de y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

plt.show()
print(eficiencia)
```
