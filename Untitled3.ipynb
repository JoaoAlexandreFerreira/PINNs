```python
import torch
import torch.autograd as autograd         #AD, diferenciação automática
from torch import Tensor
import torch.nn as nn                     #rede neural
import torch.optim as optim               #otimizadores
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.ticker
from sklearn.model_selection import train_test_split
import numpy as np
import time
import scipy.io
```

```python
#Padronizando o tipo dos valores
torch.set_default_dtype(torch.float)
torch.manual_seed(1234) #Uma semente para os valores, evitando a geração aleatória
np.random.seed(1234)

#Esse aqui não sei bem, mas sempre usam
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(device)

if device == 'cuda':
    print(torch.cuda.get_device_name())
```

```python
class RNA(nn.Module):
    def __init__(self,camadas):
        super().__init__()

        #Função de ativação
        self.ativacao = nn.Tanh()

        #Criando a rede neural
        self.linears = nn.ModuleList([nn.Linear(camadas[i], camadas[i+1]) for i in range(len(camadas)-1)])

        #inicialização normal
        for i in range(len(camadas)-1):
            nn.init.xavier_normal_(self.linears[i].weight.data, gain=1.0)


    #propagação direta
    def forward(self,x):

        if torch.is_tensor(x) != True:
            x = torch.from_numpy(x)

        u_b = torch.from_numpy(ub).float().to(device)
        l_b = torch.from_numpy(lb).float().to(device)

        #Normalizando os dados
        x = (x - l_b)/(u_b - l_b)

        a = x.float()

        for i in range(len(camadas)-2):

            z = self.linears[i](a)
            a = self.ativacao(z)

        a = self.linears[-1](a)
        return a
```

```python
#Classe da pinn para navier stokes
class PINN_NS():
  def __init__(self, camadas, espacor, espaco0, c0, espacox1, cbx1, espacox2, cbx2):

    self.x = torch.tensor(espacor[:,0:1], dtype=torch.float32, requires_grad=True).to(device)
    self.t = torch.tensor(espacor[:,1:2], dtype=torch.float32, requires_grad=True).to(device)

    self.x0 = torch.tensor(espaco0[:,0:1], dtype=torch.float32, requires_grad=True).to(device)
    self.t0 = torch.tensor(espaco0[:,1:2], dtype=torch.float32, requires_grad=True).to(device)
    self.u0 = torch.tensor(c0, dtype=torch.float32, requires_grad=True).to(device)

    self.xc = torch.tensor(espacox1[:,0:1], dtype=torch.float32, requires_grad=True).to(device)
    self.tc = torch.tensor(espacox1[:,1:2], dtype=torch.float32, requires_grad=True).to(device)
    self.uc = torch.tensor(cbx1, dtype=torch.float32, requires_grad=True).to(device)

    self.xx2 = torch.tensor(espacox2[:,0:1], dtype=torch.float32, requires_grad=True).to(device)
    self.tx2 = torch.tensor(espacox2[:,1:2], dtype=torch.float32, requires_grad=True).to(device)
    self.uc2 = torch.tensor(cbx2, dtype=torch.float32, requires_grad=True).to(device)


    self.mse = nn.MSELoss(reduction = 'mean') #Mean Squared Error, tirado do próprio pytorch
    self.itr = 0 #Contador de iteração


    self.modelo = RNA(camadas).to(device) #Instanciando a rede neural

    self.otimizador = optim.LBFGS(self.modelo.parameters(), lr=0.01, max_iter=200000, max_eval=50000,history_size=50, tolerance_grad=1e-05, tolerance_change=0.5 * np.finfo(float).eps,line_search_fn="strong_wolfe")
    #self.otimizador = optim.Adam(self.modelo.parameters(), lr=0.01)

  def EDP(self):

    u = self.modelo(torch.hstack((self.x, self.t)))

    u_x = autograd.grad(u, self.x, torch.ones_like(u), create_graph = True)[0]
    u_xx = autograd.grad(u_x, self.x, torch.ones_like(u_x), create_graph = True)[0]

    u_t = autograd.grad(u, self.t, torch.ones_like(u), create_graph = True)[0]


    #f = u_t + vel*(u_x) - difus*(u_xx + u_yy)
    f = u_t + 1.0*(u_x)

    return u, f

  def erro(self):

    u_pred, f = self.EDP()

    #Erro dos dados
    u0_pred = self.modelo(torch.hstack((self.x0, self.t0)))
    erro_u0 = self.mse(u0_pred, self.u0)

    uc_pred = self.modelo(torch.hstack((self.xc, self.tc)))
    erro_uc = self.mse(uc_pred, self.uc)
    uc2_pred = self.modelo(torch.hstack((self.xx2, self.tx2)))
    erro_uc2 = self.mse(uc2_pred, self.uc2)


    #Erro da edp
    erro_f = self.mse(f, torch.zeros_like(f))

    #Erro total
    erro_total = erro_u0 + erro_uc + erro_f + erro_uc2

    return erro_total

  def treino_otimizador(self):

    self.otimizador.zero_grad()

    erro_total = self.erro()
    erro_total.backward()

    self.itr +=1

    if self.itr % 100 == 0:
      print('Iteração: {:}, Custo: {:0.6f}'.format(self.itr, erro_total))

    return erro_total

  def treino(self):
    self.otimizador.step(self.treino_otimizador)

  def predicao(self, x, t):
    pred = self.modelo(torch.hstack((x, t)))
    return pred
```

```python
#Definindo os pontos X
N_0 = 170 #200 pontos para condição inicial
N_bx = 170 #200 pontos para condição de contorno em x
N_by = 170 #200 pontos para a condição de contorno em y
N_r = 12500 #Pontos para a edp

#Pontos do domínio
tmin = 0. ; tmax = 10. #Tempo de simulação
xmin = 0.; xmax = 10. #largura do duto


#Constantes
T = 352.89 #K
v = 2.21 #m/s
vel = 2.21
#vel = torch.from_numpy(v).float().to(device)
psimeg = 2.6 #parametro do solvente
Mmeg = 1.98+2*5.48 #Peso molecular do meg, em g/mol
mi = 0.891 #Viscosidade da solução, centipoises
Va = 55.7582 #Volume molar do soluto, em cm³/g*mol
Difu = 7.4e-8 *((np.sqrt(psimeg*Mmeg)*T)/(mi*Va**0.6))
difus = 7.4e-8 *((np.sqrt(psimeg*Mmeg)*T)/(mi*Va**0.6))
#difus = torch.from_numpy(Difu).float().to(device)
lb = np.hstack((xmin, tmin)); ub = np.hstack([xmax, tmax])
```

```python
def c_contorno(x,t):
  return np.where(x == 0, 1, 0)

def c_inicial(x):
  return np.zeros_like(x)

t0 = np.zeros((150, 1))
x0 = np.random.uniform(xmin, xmax, (150, 1))
espaco0 = np.concatenate([x0, t0], axis=1)
c0 = c_inicial(x0)

tbx1 = np.random.uniform(tmin, tmax, (150, 1))
xbx1 = np.zeros((150,1))
espacox1 = np.concatenate([xbx1, tbx1], axis=1)
cbx1 = c_contorno(xbx1, tbx1)

tbx2 = np.random.uniform(tmin, tmax, (150, 1))
xbx2 = np.ones((150,1))
espacox2 = np.concatenate([xbx2, tbx2], axis=1)
cbx2 = c_contorno(xbx2, tbx2)

xr = np.random.uniform(xmin, xmax, (N_r, 1))
tr = np.random.uniform(tmin, tmax, (N_r, 1))
espaco_r = np.concatenate([xr, tr], axis=1)
```
